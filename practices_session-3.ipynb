```python
def http_status(code):
    codes = [200, 400, 404, 500]
    messages = ["ok", "bad request", "Not found", "Internal Server Error"]

    index = codes.index(code) if code in codes else -1
    return messages[index] if index != -1 else "Unknown status"

print(http_status(200))
print(http_status(404))
```

```python
def http_status(code):
    status_map = {
        200: "ok",
        400: "bad request",
        404: "Not found",
        500: "Internal Server Error"
    }
    return status_map.get(code, "Unknown status")

print(http_status(200))
print(http_status(404))
```
