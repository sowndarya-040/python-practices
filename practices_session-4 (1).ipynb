# Print first 10 multiples of N

```python
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n * i)
```

# Print first M multiples of N

```python
n = int(input("Enter N: "))
m = int(input("Enter M: "))

for i in range(1, m + 1):
    print(n * i)
```

# Print first 10 multiples of N in reverse order

```python
n = int(input("Enter the number: "))

for i in range(10, 0, -1):
    print(n * i)
```

# Print table of userâ€™s choice

```python
num = int(input("Enter the number to print table: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```
